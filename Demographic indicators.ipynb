```python
indicators['GNI for 1 person'] = round(indicators['GNI']/indicators['Population'], 1)
indicators[['Country Name','GNI','Population','GNI for 1 person']]
```

```python
indicators[indicators['Greenhouse gas emissions'] > 150000].sort_values(by='Greenhouse gas emissions',ascending=False)[['Country Name','Region','Subregion','Intermediate Region','Greenhouse gas emissions']]
```

```python
countries_with_high_emissions = indicators[indicators['Greenhouse gas emissions'] > 150000].sort_values(by='Greenhouse gas emissions',ascending=False)[['Country Name','Greenhouse gas emissions']]
plt.figure(figsize=(20,10))
plt.bar(countries_with_high_emissions['Country Name'], countries_with_high_emissions['Greenhouse gas emissions'])
plt.title('Countries with level of greenhouse gas emissions')
plt.xlabel('Countries')
plt.ylabel('Greenhouse gas emissions')
plt.xticks(rotation = 90)
plt.show()
```

```python
GNI_by_region = indicators
GNI_by_region = GNI_by_region.groupby('Region')['GNI'].sum()
GNI_by_region = GNI_by_region.astype({'GNI': str})
GNI_by_region = GNI_by_region.apply(lambda x : x + '$')
GNI_by_region.sort_values(ascending=False)
```

```python
regions_by_GNI = indicators.groupby('Region')['GNI'].sum()
plt.figure(figsize=(20,10))
plt.bar(regions_by_GNI.index, regions_by_GNI.values)
plt.title('Regions by GNI')
plt.xlabel('Regions')
plt.ylabel('GNI')
plt.show()
```

```python
#COUNT COUNTRIES
count_country_in_regions = indicators['Region'].value_counts()
sns.barplot(x=count_country_in_regions.index, y=count_country_in_regions.values, data=indicators)
plt.xlabel("Regions")
plt.ylabel("Count country in regions")
plt.show()
```

```python
Internet_use_by_regions = indicators
Internet_use_by_regions['Internet population'] = ((Internet_use_by_regions['Internet use'] / 100) * Internet_use_by_regions['Population'])
```

```python
population_by_regions_bar_chart = indicators.groupby('Region')['Population'].sum()
sns.barplot(x=population_by_regions_bar_chart.index, y=population_by_regions_bar_chart.values, data=indicators)
plt.xlabel("Regions")
plt.ylabel("Population")
plt.show()
```

```python
sns.catplot(data=Internet_use_by_regions,
                 kind="bar",
                 x = Internet_use_by_regions['Region'],
                 y = Internet_use_by_regions['Internet population'],
                 palette = 'dark')
plt.xlabel("Regions")
plt.ylabel("Count Internet Use")
plt.show()
```

```python
#pie chart of internet users in regions
for_pie_chart = Internet_use_by_regions.groupby('Region')['Internet population'].sum().sort_values(ascending=False)
pie, ax = plt.subplots(figsize=[15,10])
labels = for_pie_chart.index
plt.pie(x=for_pie_chart, autopct="%.1f%%", labels=labels, pctdistance=0.5)
plt.title("Internet users by regions", fontsize=14);
```

```python
#9 What are the outliers of GNI, Population, tourism, Greenhouse gas emissions?
```

```python
fig = plt.figure(figsize=(16, 8))
fig.suptitle('Outliers')

ax1 = fig.add_subplot(2,2,1) 
ax2 = fig.add_subplot(2,2,2) 
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.boxplot(indicators['GNI'],sym='+', vert=False, whis=1.5)
ax2.boxplot(indicators['Population'],sym='+', vert=False, whis=1.5)
ax3.boxplot(indicators['International tourism'],sym='+', vert=False, whis=1.5)
ax4.boxplot(indicators['Greenhouse gas emissions'], sym='+', vert=False, whis=1.5)

ax1.set_title("GNI outliers")
ax2.set_title("Population outliers")
ax3.set_title("International tourism outliers")
ax4.set_title("Greenhouse gas emissions outliers")
```
